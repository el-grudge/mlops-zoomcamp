```python
import importlib

# Function to dynamically load a class
def load_class(full_class_string):
    module_path, class_name = full_class_string.rsplit('.', 1)
    module = importlib.import_module(module_path)
    return getattr(module, class_name)

# Load the LogisticRegression class dynamically
model_class_name = 'sklearn.linear_model.LogisticRegression'
mlmodel_class = load_class(model_class_name)

# Instantiate the loaded class
mlmodel_instance = mlmodel_class()

# Print the instantiated objects to verify
print(mlmodel_instance)
print(type(mlmodel_instance))
```

```python
from sklearn.linear_model import LogisticRegression
# Direct instantiation for comparison
model_class = LogisticRegression()
print(model_class)
print(type(model_class))
```

```python
model_classes = ['lightgbm.LGBMClassifier', 'sklearn.ensemble.RandomForestClassifier', 'xgboost.XGBClassifier']
models = [load_class(module_and_class_name) for module_and_class_name in model_classes]
```
